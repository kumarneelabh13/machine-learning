```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
U = []
u = np.array([1, 0])
U.append(u)

A = np.array([[.8, .3], [.2, .7]])
print(A)

for i in range(7):
    u = np.dot(A, u)
    U.append(u)

U
plt.plot(U)
plt.show()

print(np.dot(A, np.array([.6, .4])))
```
