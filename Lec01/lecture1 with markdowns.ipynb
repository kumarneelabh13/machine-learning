### Bubble sort

```python
# bubble sort
l = [2, 1, 4, 9, 7, 6]
for j in range(len(l) - 1) :
    for i in range(len(l) - 1 - j):
        if (l[i] > l[i + 1]) :
            b = l[i]
            l[i] = l[i + 1]
            l[i + 1] = b
print(l)
```

### Bubble sort as a function

```python
def bubble(l) :
    for j in range(len(l) - 1) :
        for i in range(len(l) - 1 - j):
            if (l[i] > l[i + 1]) :
                b = l[i]
                l[i] = l[i + 1]
                l[i + 1] = b
                
                
l = [4,3,2,1]
bubble(l)
print(l)
```
