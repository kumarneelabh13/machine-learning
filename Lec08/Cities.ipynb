```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cities = pd.read_csv('cities.csv')
```

```python
cities.head()
print(cities.columns)
df = cities[['effective_literacy_rate_female', 'effective_literacy_rate_male', 'child_sex_ratio']].copy()
```

```python
import seaborn as sns
def plot_correlation_map( df ):
    corr = df.corr()
    _ , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    _ = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
    )
```

```python
plot_correlation_map(df)
df.corr()
df
```
